```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib
```

```python
symptomsDf = pd.read_csv(r"./DiseaseDiagnose/DiseaseSymptoms.csv")
symptomsDf.columns.tolist()
```

```python
symptomsDf.head()
has_na = symptomsDf.isna()
# print(has_na)
```

```python
symptomsDf = symptomsDf.drop("Unnamed: 133", axis=1)
symptomsDf.head()
```

```python
diagnosisTestDf = pd.read_csv("./DiseaseDiagnose/Testing.csv")
diagnosisTestDf.head()
```

```python
X_train = symptomsDf.drop("prognosis", axis=1)
Y_train = symptomsDf["prognosis"]

X_test = diagnosisTestDf.drop("prognosis", axis=1)
Y_test = diagnosisTestDf["prognosis"]

# model = DecisionTreeClassifier()
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)
report = classification_report(Y_test, Y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report: \n", report)
joblib.dump(model, 'trained_model.joblib')
```
